```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matplotlib inline
```

```python
df = pd.read_csv('5yo_startups_cleaned.csv')
```

```python
from datetime import datetime

active_startups = df.drop(df.index[df.stage == 'Exited'])
age = (pd.to_datetime(datetime(2015, 1, 1)) - pd.to_datetime(active_startups.est_founding_date)).astype('timedelta64[D]')
```

```python
toddlers = active_startups[age > (365*2)].copy()
```

```python
toddlers.stage.value_counts()
```

```python
got_funded = toddlers[toddlers.stage.isin(['A', 'B', 'C', 'Late']) & (toddlers['pre-series-A funding'] > 0)]['pre-series-A funding'].apply(math.log10).values
not_funded_yet = toddlers[(toddlers.stage == u'Pre Series A') & (toddlers['pre-series-A funding'] > 0)]['pre-series-A funding'].apply(math.log10).values

plt.style.use('fivethirtyeight')
plt.hist(not_funded_yet, label="Not series A-funded", histtype="stepfilled", bins=np.arange(3, 8, 0.25), alpha=0.8)
plt.hist(got_funded, label="Series A-funded", histtype="stepfilled", bins=np.arange(3, 8, 0.25), alpha=0.4)
plt.legend(loc="upper left")
plt.xlabel('Total pre-series-A funding (log scale)')
plt.ylabel('# startups')
plt.savefig('total_pre_series_A_funding.png', dpi=400, bbox_inches='tight')
```

```python
got_funded = toddlers[toddlers.stage.isin(['A', 'B', 'C', 'Late'])][u'expansion rate'].values * 365
not_funded_yet = toddlers[toddlers.stage == u'Pre Series A'][u'expansion rate'].values * 365

plt.style.use('ggplot')
plt.hist(got_funded, label="Series A-funded", histtype="stepfilled", bins=np.arange(0, 24, 2), alpha=0.8)
plt.hist(not_funded_yet, label="Not series A-funded", histtype="stepfilled", bins=np.arange(0, 24, 2), alpha=0.4)
plt.legend(loc="upper right")
plt.xlabel('Expansion rate')
plt.ylabel('# startups')
plt.savefig('expansion_rate.png', dpi=400, bbox_inches='tight')
```

```python
got_funded = toddlers[toddlers.stage.isin(['A', 'B', 'C', 'Late']) & (toddlers['pre-series-A funding'] > 0)].copy()
not_funded_yet = toddlers[(toddlers.stage == u'Pre Series A') & (toddlers['pre-series-A funding'] > 0)].copy()

plt.style.use('ggplot')
plt.plot(got_funded['expansion rate'].values * 365, got_funded['pre-series-A funding'], 'o', label="Series A-funded", alpha=0.8)
plt.plot(not_funded_yet['expansion rate'].values * 365, not_funded_yet['pre-series-A funding'], 'o', label="Not series A-funded", alpha=0.4)
plt.xlim((-5, 40))
plt.ylim((10 ** 2.5, 10 ** 8))
plt.semilogy()
plt.legend(loc='lower right')
plt.xlabel('Expansion rate')
plt.ylabel('Total pre-series A funding (USD)')
plt.savefig('feature_interaction.png', dpi=400, bbox_inches='tight')
```
