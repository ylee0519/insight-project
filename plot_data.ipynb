```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
startups = pd.read_csv('5yo_startups_cleaned.csv')
```

```python
matplotlib inline
```

```python
startups.stage.value_counts()
```

```python
got_funded = startups[startups.stage.isin(['A', 'B', 'C', 'Late']) & (startups.total_funding > 0)][u'total_funding'].apply(math.log10).values
not_funded_yet = startups[(startups.stage == u'Pre Series A') & (startups.total_funding > 0)][u'total_funding'].apply(math.log10).values

plt.style.use('fivethirtyeight')
plt.hist(not_funded_yet, label="Not funded (yet)", histtype="stepfilled", bins=20, alpha=0.8)
plt.hist(got_funded, label="Got funded", histtype="stepfilled", bins=20, alpha=0.8)
plt.legend(loc="upper left")
plt.xlabel('Total funding (log scale)')
plt.ylabel('# 5yo startups')
plt.savefig('total_funding.png', dpi=400, bbox_inches='tight')
```

```python
got_funded = startups[startups.stage.isin(['A', 'B', 'C', 'Late']) & (startups.seed_funding > 0)][u'seed_funding'].apply(math.log10).values
not_funded_yet = startups[(startups.stage == u'Pre Series A') & (startups.seed_funding > 0)][u'seed_funding'].apply(math.log10).values

plt.hist(not_funded_yet, label="Not funded (yet)", histtype="stepfilled", bins=20, alpha=0.8)
plt.hist(got_funded, label="Got funded", histtype="stepfilled", bins=20, alpha=0.8)
plt.legend(loc="upper left")
plt.xlabel('Total pre-series A funding (log scale)')
plt.ylabel('# 5yo startups')
plt.savefig('total_seed_funding.png', dpi=400, bbox_inches='tight')
```

```python
got_funded = startups[startups.stage.isin(['A', 'B', 'C', 'Late']) & (startups.employee > 0)][u'employee'].apply(math.log10).values
not_funded_yet = startups[(startups.stage == u'Pre Series A') & (startups.employee > 0)][u'employee'].apply(math.log10).values

plt.hist(not_funded_yet, label="Not funded (yet)", histtype="stepfilled", bins=np.arange(0, 4, 0.25), alpha=0.8)
plt.hist(got_funded, label="Got funded", histtype="stepfilled", bins=np.arange(0, 4, 0.25), alpha=0.8)
plt.legend(loc="upper right")
plt.xlabel('Number of employees (log scale)')
plt.ylabel('# 5yo startups')
```

```python
the_fundeds = startups[startups.stage.isin(['A', 'B', 'C', 'Late']) & (startups.total_funding > 0) & (startups.employee > 0)].copy()
the_unfortunates = startups[(startups.stage == u'Pre Series A') & (startups.total_funding > 0) & (startups.employee > 0)].copy()

plt.scatter(the_fundeds['employee'], the_fundeds['total_funding'], label="Got funded", alpha=0.2, c='red')
plt.scatter(the_unfortunates['employee'], the_unfortunates['total_funding'], label="Not funded (yet)", alpha=0.2, c='blue')
plt.xlim((10, 1000))
plt.loglog()
plt.legend(loc="lower right")
plt.xlabel('Employee number')
plt.ylabel('Total funding')
```
